# Net - experiments

## LEARNIG RATE

```python
import glob

# get data file names
path =r'..\data\results\Net\bs8'
filenames = glob.glob(path + "/*.csv")

dfs = []
for filename in filenames:
    dfs.append(pd.read_csv(filename))
```

```python
net_lr_test_acc = dfs[0].drop("Unnamed: 0", axis =1)
net_lr_test_loss = dfs[1].drop("Unnamed: 0", axis =1)
net_lr_train_acc = dfs[2].drop("Unnamed: 0", axis =1)
net_lr_train_loss = dfs[3].drop("Unnamed: 0", axis =1)
```

```python
plt.plot(net_lr_test_acc, alpha = 0.7)
plt.legend(["test " + lr for lr in net_lr_test_acc.columns])
plt.title("Net - Accuracy")
```

```python
plt.plot(net_lr_train_acc, alpha = 0.7)
plt.legend(["train " + lr for lr in net_lr_train_acc.columns])
plt.title("Net - Accuracy")
```

```python
plt.plot(net_lr_test_loss, alpha = 0.7)
plt.legend(["test " + lr for lr in net_lr_test_loss.columns])
plt.title("Net - Loss")
plt.show()

plt.plot(net_lr_train_loss, alpha = 0.7)
plt.legend(["train " + lr for lr in net_lr_train_loss.columns])
plt.title("Net - Loss")
plt.show()
```

# RESNET_2

```python
test_acc = pd.read_csv("..\\data\\results\\resnet18_2\\SGD_ep30_lr-3\\batch_test_acc_history.csv")
train_acc = pd.read_csv("..\\data\\results\\resnet18_2\\SGD_ep30_lr-3\\batch_train_acc_history.csv")
test_loss = pd.read_csv("..\\data\\results\\resnet18_2\\SGD_ep30_lr-3\\batch_test_loss_history.csv")
train_loss = pd.read_csv("..\\data\\results\\resnet18_2\\SGD_ep30_lr-3\\batch_train_loss_history.csv")

test_acc.drop("Unnamed: 0", axis = 1, inplace = True)
train_acc.drop("Unnamed: 0", axis = 1, inplace = True)
test_loss.drop("Unnamed: 0", axis = 1, inplace = True)
train_loss.drop("Unnamed: 0", axis = 1, inplace = True)
test_acc.drop("Unnamed: 0.1", axis = 1, inplace = True)
train_acc.drop("Unnamed: 0.1", axis = 1, inplace = True)
test_loss.drop("Unnamed: 0.1", axis = 1, inplace = True)
train_loss.drop("Unnamed: 0.1", axis = 1, inplace = True)


plt.plot(test_acc)
plt.legend(test_acc.columns)
plt.title("Resnet18 Last layer trained -  Accuracy - Test")
plt.show()


plt.plot(train_acc)
plt.legend(train_acc.columns)
plt.title("Resnet18 Last layer trained -  Accuracy - Train")
plt.show()

plt.plot(test_loss)
plt.legend(test_loss.columns)
plt.title("Resnet18 Last layer trained -  Loss - Test")
plt.show()


plt.plot(train_loss)
plt.legend(train_loss.columns)
plt.title("Resnet18 Last layer trained -  Loss - Train")
plt.show()
```
